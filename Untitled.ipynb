```python
import os
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
files = []
for filename in os.listdir('swcm_logs'):
    if filename.startswith('access'):
        files.append(filename)
```

```python
matches = ["Zabbix", "docs", "health", "debug", "metrics", "stats", "openapi", "v1", "v3", "favicon"]

rows = []
for filename in files:
    print(f"Opening file {filename}")
    with open(f"swcm_logs/{filename}", 'r') as logfile:
        for line in logfile:
            if any(x in line for x in matches):
                continue
            try:
                line = line.split()
                date = datetime.strptime(line[3][1:], '%d/%b/%Y:%H:%M:%S')
                tz = line[4][:-1]
                ip = line[0]
                method = line[5][1:]
                url = line[6]
                statusCode = line[8]
                responseTime = line[14][5:-1]
                rows.append((date, str(ip), str(method), str(url), int(statusCode), responseTime))
            except Exception:
                print(line)
                continue


df = df.from_records(rows, columns=['Time', 'IP', 'HttpMethod', 'URL', 'StatusCode', 'ResponseTime'])

        
```

```python
df['Date'] = df.apply(lambda x: x['Time'].date(), axis=1)
```

```python
def filterRt(row):
    if row['ResponseTime'] == '-':
        return float(0)
    else:
        return float(row['ResponseTime'])

df['URT'] = df.apply(filterRt, axis=1)
```

```python
df['Module'] = df.apply(lambda x: x['URL'].split('/')[1], axis=1)
```

```python
df200 = df.loc[df['StatusCode'] == 200]
```

```python
df200.plot(y='URT', x='Time', kind='scatter', figsize=(16, 8), title='Response Time Distribution')
plt.show()
```

```python
modules = np.unique(df200.Module)
colors = [plt.cm.tab10(i/float(len(modules)+3)) for i in range(len(modules))]

# Draw Plot for Each Module
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k');

for i, module in enumerate(modules):
    plt.scatter('Time', 'URT', 
                data=df200.loc[df200.Module==module, :], 
                s=20, color=colors[i], label=str(module));

# Decorations
plt.gca().set(xlabel='Time', ylabel='URT');

plt.xticks(fontsize=12); plt.yticks(fontsize=12);
plt.title("Response Time Distribution by Module", fontsize=22);
plt.legend(fontsize=12);   
plt.show()  
```

```python
def meanRTbyModules(df):
    res = {}
    modules = ['port', 'port_vlan', 'system']
    for module in modules:
        res[module] = df.loc[df.Module==module, :].mean()
    res['common'] = df.mean()
    return res
```
